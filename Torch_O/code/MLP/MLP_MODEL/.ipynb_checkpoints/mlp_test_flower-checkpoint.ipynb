```python
EPOCH = 40
BATCH_SIZE=30
NUM_WORKERS=0
LEARNING_RATE = 0.00005
```

```python
#SGD
fd = FlowersDataset()
fm = Mlp_Torch('flowers_model_1', fd, [12])
fm.exec_all(epoch_count=EPOCH, report=4, batch_size =BATCH_SIZE \
           , learning_rate = LEARNING_RATE , num_workers = NUM_WORKERS \
          )
```

```python
# Adam
fd = FlowersDataset()
fm = Mlp_Torch('flowers_model_1', fd, [12],use_adam=True)
fm.exec_all(epoch_count=EPOCH, report=4, batch_size =BATCH_SIZE \
           , learning_rate = LEARNING_RATE , num_workers = NUM_WORKERS)
```

```python
#activate + adam
fd = FlowersDataset()
fm = Mlp_Torch('flowers_model_1', fd, [40,30,10],use_adam=True,auto_activate = True)
fm.exec_all(epoch_count=EPOCH, report=4, batch_size =BATCH_SIZE \
           , learning_rate = LEARNING_RATE , num_workers = NUM_WORKERS)
```
